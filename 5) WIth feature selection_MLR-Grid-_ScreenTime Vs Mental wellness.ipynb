```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.metrics import r2_score
```

```python

def selectkbest(indep_x2,dep_y1,n):
        test = SelectKBest(score_func=chi2, k=n)
        fit1= test.fit(indep_x2,dep_y1)
        # summarize scores       
        selectk_features = fit1.transform(indep_x2)
        return selectk_features
def split_scalar(indep_X2,dep_Y1): 
        from sklearn.preprocessing import StandardScaler
        from sklearn.model_selection import train_test_split 
        X_train, X_test, y_train, y_test = train_test_split(indep_x2, dep_y1, test_size = 0.25, random_state = 0)
        sc = StandardScaler()
        X_train_pp = sc.fit_transform(X_train)
        X_test_pp = sc.transform(X_test)    
        return X_train_pp, X_test_pp, y_train, y_test
def mlr(x_train_pp,x_test_pp,y_train,y_test):    
    from sklearn.linear_model import LinearRegression
    reg=LinearRegression()
    reg.fit(x_train_pp,y_train)
    y_pred = reg.predict(x_test_pp)
    re=r2_score(y_test,y_pred)
    return re
    
```

```python
df=pd.read_csv('ScreenTime_vs_MentalWellness.csv')
df = df.drop(['Unnamed: 15','user_id'],axis=1)
df=pd.get_dummies(df,drop_first=True).astype(int)
df.isnull().sum()
```

```python
indep_x1=df[['age', 'screen_time_hours', 'work_screen_hours', 'leisure_screen_hours',
       'sleep_hours', 'sleep_quality_1_5', 'stress_level_0_10',
       'productivity_0_100', 'exercise_minutes_per_week',
       'social_hours_per_week', 'gender_Male',
       'gender_Non-binary/Other', 'occupation_Retired',
       'occupation_Self-employed', 'occupation_Student',
       'occupation_Unemployed', 'work_mode_In-person', 'work_mode_Remote']]
dep_y1=df['mental_wellness_index_0_100']
```

```python
indep_x2 = pd.get_dummies(indep_x1)
```

```python
kbest=selectkbest(indep_x2,dep_y1,2)   
x_train_pp,x_test_pp,y_train,y_test=split_scalar(kbest,dep_y1)
re1mlr=mlr(x_train_pp,x_test_pp,y_train,y_test)
result=re1mlr
```
